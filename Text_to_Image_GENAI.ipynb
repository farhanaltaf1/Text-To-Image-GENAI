```python
from diffusers import DiffusionPipeline
```

```python
import torch
pipe=DiffusionPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-base-1.0", ##model
    torch_dtype=torch.torch.float16,##image is in array form it will convert image pixel values to float type
    USE_SAFETENSORS=True, ##it will reduce numeric error in image generation
    variant='fp16'##model is using 16bit floating as mentioned in datatype also called half precision

)

pipe.to("cuda")
```

```python
propmt="a boy sitting on table using computer"

negative_propmpt="avoid including other people in the scene "

Image=pipe(prompt=propmt,negative_prompt=negative_propmpt,).images[0] #dont include latent
```

```python
import matplotlib.pyplot as plt #matplotlib library because it will be use graph to display image


plt.imshow(Image)
plt.axis('off')
plt.show()
```

```python
def generate_image(propmt,negative_propmpt):
  Image=pipe(prompt=propmt,negative_prompt=negative_propmpt,).images[0]
  plt.imshow(Image)
  plt.axis('off')
  plt.show()
```

```python
propmt="elon musk holding pakistan flag " #type anything im this
negative_propmpt="avoid including other people in scene"
generate_image(propmt,negative_propmpt)
```
